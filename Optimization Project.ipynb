```julia
using CSV, DataFrames
data = DataFrame(CSV.File("C:\\Users\\tfurr\\OneDrive\\Documents\\optprojectdata.csv"))
```

```julia
data = data[:,1:33]
data = data[completecases(data), :]
```

```julia
df = data[:, [:PlayerID, :Player, :Pos, :PER, :X2022_23, :C, :SG, :SF, :PG, :PF, :WS]]
```

```julia
df2 = unique(df)
```

```julia
using JuMP, GLPK;

model = Model(GLPK.Optimizer)
n = 20
c = df2[1:n, 4]
a = df2[1:n, 5]
mat = df2[1:n ,6:10]
@variable(model, x[i=1:n], Bin)
@constraint(model, sum(a[i]*x[i] for i = 1:n) <= 134000000)
@constraint(model, sum(x[i] for i=1:n) <= 15)
@constraint(model, sum(x[i] for i=1:n) >= 14)
for j = 1:5
    @constraint(model, sum(x[i]*mat[i, j] for i=1:n) >= 2)
end
@objective(model, Max, sum(c[j]*x[j] for j=1:n))
print(model)
```

```julia
JuMP.optimize!(model)

println("Objective value: ", JuMP.objective_value(model))
println("Optimal solution is x = \n", JuMP.value.(x))
```
